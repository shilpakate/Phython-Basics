```python
from bs4 import BeautifulSoup
import requests
url =("https://www.bollywoodmdb.com/movies/external-reviews/hindi-medium-2/16445")
res = requests.get(url) 
soup = BeautifulSoup(res.text,"html.parser") 
print("*******************************************************************")
print(soup)
```

```python

reviews=soup.findAll('div', {'class':'critics-container'})
print(type(reviews))
for i in range(0,5):
   reviews[i]=reviews[i].text
   print("*************next***********************************************")
   print(reviews[i])
```
